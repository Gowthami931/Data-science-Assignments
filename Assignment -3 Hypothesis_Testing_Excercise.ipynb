Hypothesis Testing Exercise

1. A F&B manager wants to determine whether there is any significant difference in the diameter of the cutlet between two units. A randomly selected sample of cutlets was collected from both units and measured? Analyze the data and draw inferences at 5% significance level. Please state the assumptions and tests that you carried out to check validity of the assumptions.


     Minitab File : Cutlets.mtw


```python
from google.colab import files
upload=files.upload()
```

```python
import pandas as pd
import numpy as np
df=pd.read_csv("Cutlets.csv")
df
```

```python
import scipy.stats as stats
Zcal,Pvalue=stats.ttest_ind(df["Unit A"],df["Unit B"])
print("Zcalculated value:",Zcal.round(4))
print("p-value:",Pvalue.round(4))
```

```python
alpha=0.05
if Pvalue < alpha:
  print("There is a significant difference in the diameter of the cutlet between two units")
else:
  print("There is no significant difference in the diameter of the cutlet between two units")
```

2.A hospital wants to determine whether there is any difference in the average Turn Around Time (TAT) of reports of the laboratories on their preferred list. They collected a random sample and recorded TAT for reports of 4 laboratories. TAT is defined as sample collected to report dispatch.
   
  Analyze the data and determine whether there is any difference in average TAT among the different laboratories at 5% significance level.


    Minitab File: LabTAT.mtw


```python
from google.colab import files
upload=files.upload()
```

```python
import pandas as pd
import numpy as np
df=pd.read_csv("LabTAT.csv")
df
```

```python
lab1 = df["Laboratory 1"]
lab2 = df["Laboratory 2"]
lab3 = df["Laboratory 3"]
lab4 = df["Laboratory 4"]
```

```python
import scipy.stats as stats
Fcal,pval=stats.f_oneway(lab1,lab2,lab3,lab4)
```

```python
Fcal,pval
```

```python
print("F calculated value:",Fcal.round(4))
print("p-value:",pval.round(4))
```

```python
alpha=0.05
if pval < alpha:
  print("There is significant difference in average TAT among the different Labratories")
else:
  print("There is no significant difference in average TAT among the difference labratories")
```
