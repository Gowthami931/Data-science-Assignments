```python
import numpy as np
import pandas as pd
df=pd.read_csv("delivery_time.csv")
df
```

```python
df.corr()
```

```python
x=df[["Sorting Time"]]
x
```

```python
y=df[["Delivery Time"]]
y
```

```python
import matplotlib.pyplot as plt
plt.scatter(x,y)
```

```python
from sklearn.linear_model import LinearRegression
LR = LinearRegression()


```

```python
LR.fit(x,y)
```

```python
LR.intercept_
```

```python
LR.coef_
```

```python
y_pred = LR.predict(x)
y_pred
```

```python
plt.scatter(x,y,color="blue")
plt.scatter(x,y_pred,color="black")
plt.plot(x,y_pred,color="red")
plt.show()
```

```python
from sklearn.metrics import mean_squared_error
MSE=mean_squared_error(y,y_pred)
print("mean square error:",MSE.round(3))
print("root mean square error;",np.sqrt(MSE).round(3))
```

```python
xlog1=np.sqrt(x)
y_pred_transform=LR.predict(xlog1)
y_pred_transform
from sklearn.metrics import mean_squared_log_error
mse=mean_squared_error(y,y_pred_transform)
print("mean square error:",mse.round(4))
print("root mean square error:",np.sqrt(mse).round(4))
```

```python
xlog2=np.cbrt(x)
y_pred_transform=LR.predict(xlog2)
y_pred_transform
```

```python
from sklearn.metrics import mean_squared_error
mse=mean_squared_error(y,y_pred_transform)
print("mean square error :",mse.round(4))
print("root mean square error:",np.sqrt(mse).round(4))
```

```python
import pandas as pd
import numpy as np
df=pd.read_csv("Salary_Data.csv")
df
```

```python
df.corr()
```

```python
x=df[["YearsExperience"]]
x
```

```python
y=df[["Salary"]]
y
```

```python
import matplotlib.pyplot as plt
plt.scatter(x,y)
```

```python
from sklearn.linear_model import LinearRegression
LR=LinearRegression()
LR.fit(x,y)
LR.intercept_
LR.coef_
y_pred=LR.predict(x)
y_pred
```

```python
import matplotlib.pyplot as plt
plt.scatter(x,y,color="blue")
plt.scatter(x,y_pred,color="red")
plt.plot(x,y_pred,color="black")
plt.show()
```

```python
from sklearn.metrics import mean_squared_error
MSE=mean_squared_error(y,y_pred)
print("mean square error:",MSE.round(3))
print("root mesn square error:",np.sqrt(MSE).round(3))
```

```python
xlog1=np.sqrt(x)
y_pred_transform=LR.predict(xlog1)
y_pred_transform
from sklearn.metrics import mean_squared_error
mse=mean_squared_error(y,y_pred_transform)
print("mean square error:",mse.round(4))
print("root mean square error:",np.sqrt(mse).round(4))
```

```python
xlog2=np.cbrt(x)
y_pred_transform=LR.predict(xlog2)
y_pred_transform
from sklearn.metrics import mean_squared_error
mse=mean_squared_error(y,y_pred_transform)
print("mean square error:",mse.round(4))
print("root mean square error:",np.sqrt(mse).round(4))
```
